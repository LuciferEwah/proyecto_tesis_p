```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
import numpy as np
import pandas as pd
# Simular un DataFrame con datos ficticios
np.random.seed(0)
df_simulado = pd.DataFrame({
    'Antecedente1': np.random.choice(['1','0'],100000),
    'Antecedente2': np.random.choice(['1','0'],100000),
    'Antecedente3': np.random.choice(['1','0'],100000),
    'Medida1': np.random.choice(['1','0'], 100000),
    'Medida2': np.random.choice(['1','0'], 100000),
    'Medida3': np.random.choice(['1','0'], 100000)
})

# Combinar las medidas en una sola columna (ejemplo simplificado)
df_simulado['Medida_Combinada'] = df_simulado[['Medida1', 'Medida2','Medida3']].apply(lambda x: '-'.join(x), axis=1)

# Preparar los datos para el modelo
X = df_simulado[['Antecedente1', 'Antecedente2', 'Antecedente3']]  # Características
y = df_simulado['Medida_Combinada']                                # Etiqueta objetivo

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=100)

# Crear y entrenar el modelo de árbol de decisión
modelo = DecisionTreeClassifier(random_state=1)
modelo.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred = modelo.predict(X_test)

# Calcular la precisión
precision = accuracy_score(y_test, y_pred)
precision
```

```python
# Ya que el código proporcionado entrena el modelo y hace predicciones, ahora procederé a graficar el árbol de decisión.

plt.figure(figsize=(20,10))
plot_tree(modelo, filled=True, feature_names=['Antecedente1', 'Antecedente2', 'Antecedente3'], class_names=modelo.classes_)
plt.show()
```
